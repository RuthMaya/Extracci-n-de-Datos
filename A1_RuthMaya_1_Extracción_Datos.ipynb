**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion, obteniendo los productos de los que se fabricaron más de 20 unidades
filtro6=df[df["CANT"] > 20]
filtro6
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y, en donde se muestren los registros que cuenten con una cantidad mayor a 1 y que sean del público en general
filtro7=df[(df["CANT"] > 3) & (df["NOMBRE.1"] == "PÚBLICO EN GENERAL")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro7_A1.csv")
#descargar archivo filtrado en csv
from google.colab import files 
files.download("Filtro7_A1.csv")
```

```python
#Filtro o, en donde muestra las cantidades menores a 2 o los valores que tengan como nombre el público general
filtro8=df[(df["CANT"] < 2) | (df["NOMBRE.1"] == "PÚBLICO EN GENERAL")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro8_A1.csv")
#descargar archivo filtrado en csv
from google.colab import files 
files.download("Filtro8_A1.csv")
```

```python
#Filtro not, donde muestra los valores que no son mayores a 2 y que no son del público en general
filtro9=df[~(df["CANT"] >= 2) & ~(df["NOMBRE.1"] == "PÚBLICO EN GENERAL")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9_A1.csv")
#descargar archivo filtrado en csv
from google.colab import files 
files.download("Filtro9_A1.csv")
```
